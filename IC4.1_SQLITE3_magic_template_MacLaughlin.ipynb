```python
%%sql
CREATE TABLE products (product_id INTEGER PRIMARY KEY, name TEXT, price REAL) STRICT;
```

```python
%%sql
INSERT INTO products (name, price) VALUES ('Superbowl Jersey', 120.00);
INSERT INTO products (name, price) VALUES ('Team Cap', 25.50);
INSERT INTO products (name, price) VALUES ('Fan Scarf', 30.00);
```

```python
%%sql
SELECT * FROM products;
```

```python
%%sql
UPDATE products SET price=99.99 WHERE name = 'Superbowl Jersey';
```

```python
%%sql
DELETE FROM products WHERE product_id=1;
```

```python
%%sql
SELECT * FROM products WHERE price > 50;
```
